# How to install Python using Conda

1. Download the Python3.7 Graphical installer from the conda website: https://www.anaconda.com/distribution/
2. Follow the conda installation instructions

# How to install the SDK



### 1. Open up an Anaconda Prompt

Hit the start button and type anaconda prompt.



### 2. Use pip to install the sdk

Run `pip install --user vortexasdk` in the anaconda command prompt



You're done! The VortexaSDK has now been installed.

# How to set the VORTEXA_API_KEY environment variable

### 1. Hit the windows key, then type "environment" to open up a control panel settings page titled "Edit the system environment variables"



### 2. In the System Properties window, click on the Advanced tab, then click the Environment Variables button near the bottom of that tab.

### 3. Add a new user variable



That's it, you've added a new environment variable!
